# Wildfire Preprocessing Pipeline
This notebook shows how to use the production-grade preprocessing utilities.
It covers validation, missing values, outliers, scaling, encoding, feature selection, and splitting.

```python
import pandas as pd
from app.ml.preprocessing import PreprocessConfig, preprocess_dataset
from app.ml.validation import RangeRule, SchemaSpec
```

```python
df = pd.read_csv("data/wildfire_features.csv")
df.head()
```

```python
config = PreprocessConfig(
    numeric_features=["temperature", "humidity", "rainfall", "wind_speed", "vegetation", "soil_moisture"],
    categorical_features=[],
    target_column="fire_occurrence",
    test_size=0.2,
    random_state=42,
    k_best=6,
)
```

```python
schema_spec = SchemaSpec(
    required_columns=["temperature", "humidity", "rainfall", "wind_speed", "vegetation", "soil_moisture", "fire_occurrence"],
    target_column="fire_occurrence",
)
```

```python
range_rules = [
    RangeRule(column="temperature", min_value=-60, max_value=60),
    RangeRule(column="humidity", min_value=0, max_value=100),
    RangeRule(column="rainfall", min_value=0),
    RangeRule(column="wind_speed", min_value=0),
    RangeRule(column="vegetation", min_value=0, max_value=1),
    RangeRule(column="soil_moisture", min_value=0, max_value=1),
]
```

```python
X_train, X_test, y_train, y_test, pipeline = preprocess_dataset(
    df,
    config,
    schema_spec,
    range_rules,
)
X_train.shape, X_test.shape
```
